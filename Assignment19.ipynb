Q1. Explain GET and POST methods.

```python
"""GET is used to retrieve data from a server,
   POST is used to send data to a server for creating 
   or updating a resource. GET is visible in the URL, POST is not."""
```

Q2. Why is request used in Flask?

```python
""" request is used to handle incoming HTTP requests from clients. 
    It allows the Flask application to access and process data sent
    by the client, such as form data, query parameters, and request headers."""
```

Q3. Why is redirect() used in Flask?

```python
"""redirect() in Flask is used to send a redirect response to the client's browser,
   directing it to a different URL within the application, providing navigation or 
   handling form submissions."""
```

Q4. What are templates in Flask? Why is the render_template() function used?

```python
"""Templates in Flask are files that contain HTML, along with dynamic 
   placeholders for data. The render_template() function is used to render
   these templates, replacing the placeholders with actual data, and generating
   a complete HTML response to be sent back to the client."""
```

```python
from flask import Flask,request,jsonify
app=Flask(__name__)
@app.route("/postman_data",methods=["GET","POST"])
def square_add():
    if request.method=="POST":
        num1=request.json['a']
        num2=request.json['b']
        result=num1**2 + num2**2
    return jsonify(result)
if __name__=='__main__':
    app.run(host="0.0.0.0")
```
