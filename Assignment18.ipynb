Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

```python
"""The url_for() function is used in Flask for URL building. 
   It takes the name of a view function as its first argument, 
   and any number of keyword arguments as its subsequent arguments. 
   The keyword arguments are used to dynamically generate the URL."""


from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, Flask!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index'))
        print(url_for('profile', username='shivam'))
```
